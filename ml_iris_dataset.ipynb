```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score,mean_squared_error, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('/content/iris_data.csv')
df.head()
```

```python
features = df.drop(['petal_width','species'],'columns')
features.head()
```

```python

km = KMeans(n_clusters=3)
km.fit(features)
labels=km.labels_
centers=km.cluster_centers_
```

```python
features['classes']=labels
features.head()
```

```python
df0=features[features['classes']==0]
df1=features[features['classes']==1]
df2=features[features['classes']==2]
```

```python
labels = df['petal_width']
labels.head()
```

```python
pca = PCA(n_components=2)
features1=pca.fit_transform(features)
features1
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,labels,test_size=0.2)
```

```python
print(len(features1))
print(len(x_train))
print(len(x_test))
```

```python
from sklearn.neighbors import KNeighborsRegressor
clf1 = KNeighborsRegressor(n_neighbors=3)
clf1.fit(x_train, y_train)
y_pred1 = clf1.predict(x_test)
```

```python
print("mean square error: ", mean_squared_error(y_pred1, y_test))
```

```python
from sklearn.linear_model import LinearRegression
clf3 = LinearRegression().fit(x_train, y_train)
y_pred3 = clf3.predict(x_test)
```

```python
print("mean square error for linear regression: ", mean_squared_error(y_pred3, y_test))
```
